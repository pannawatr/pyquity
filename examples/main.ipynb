```python
import pyquity
import osmnx as ox
import numpy as np
import matplotlib.pyplot as plt
```

## Calgary, Canada - With 1000 Scooters, Education, Grid size 2500x2500 meters, Served in 15 minutes

```python
place_name = 'Calgary, Canada'
G_walk = pyquity.graph_from_place(place_name, network_type='walk')
G_bike = pyquity.graph_from_place(place_name, network_type='bike')
G_gtfs = pyquity.graph_from_gtfs('/Users/net/Downloads/calgary.zip')
```

```python
MG_walk = pyquity.multimodal_graph(G_walk, G_gtfs)
MG_bike = pyquity.multimodal_graph(G_bike, G_gtfs)
```

```python
amenity = pyquity.amenity_from_place(place_name, amenity_type="education")

# Grid size 2500x2500m
grid = pyquity.grid_from_place(place_name, grid_size=2500)
grid = pyquity.amenity_in_grid(grid, amenity)
grid = pyquity.micromobility_in_grid(grid, amenity, micromobility_size=1000)
```

```python
equity = pyquity.Equity(MG_walk, MG_bike, grid, amenity)
```

```python
grid = equity.sufficientarianism(served_time=15)
gini, lorenz = equity.egalitarianism(grid)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
grid.plot(ax=ax, column='served', cmap='Reds', edgecolor='black', legend=True)
ax.set_axis_off()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
grid.plot(ax=ax, column='amenity_count', cmap='Reds', edgecolor='black', legend=True)
ax.set_axis_off()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

x = np.linspace(0, 1, len(lorenz))
ax.plot(x, lorenz, label='Lorenz Curve', color='blue')
ax.plot([0, 1], [0, 1], label='Line of Equality', color='red', linestyle='--')
ax.set_title(f"Lorenz Curve (Gini = {gini:.3f})", fontsize=14)

ax.legend()

ax.set_xlabel("Cumulative share of population")
ax.set_ylabel("Cumulative share of income")

plt.grid(True, linestyle=':', alpha=0.5)
plt.show()
```
