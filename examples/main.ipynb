```python
import pyquity
import osmnx as ox
import matplotlib.pyplot as plt
```

## Calgary, Canada

```python
place_name = 'Calgary, Canada'
G_walk = pyquity.graph_from_place(place_name, network_type='walk')
G_bike = pyquity.graph_from_place(place_name, network_type='bike')
G_gtfs = pyquity.graph_from_gtfs('/Users/net/Downloads/calgary.zip')
```

```python
MG_walk = pyquity.multimodal_graph(G_walk, G_gtfs)
MG_bike = pyquity.multimodal_graph(G_bike, G_gtfs)
```

```python
amenity = pyquity.amenity_from_place(place_name, amenity_type="education")

grid = pyquity.grid_from_place(place_name, grid_size=1500)
grid = pyquity.amenity_in_grid(grid, amenity)
grid = pyquity.micromobility_in_grid(grid, amenity, micromobility_size=1000)
```

```python
equity = pyquity.Equity(MG_walk, MG_bike, grid, amenity)
```

```python
grid = equity.sufficientarianism(served_time=15)
gini, lorenz = equity.egalitarianism(grid)
```

```python
grid.plot(column='served', cmap='Reds', edgecolor='black')
plt.show()
```

```python
grid.plot(column='amenity_count', cmap='Reds', edgecolor='black')
plt.show()
```
